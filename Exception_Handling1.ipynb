```python
 """ Errors defined in the Arithmetic Error class are 1.Zero division error
                                                 2.overrflow error
                                                 3.floating point error """
#zero division error
#it occurs when the code requires a number to be divisible by zero
try:
    a = 11
    c = a / 0
    print(c)
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")

#overflow error
#it occurs when result of a operation exceeds maximum limit
j = 5.0

try:
    for i in range(1, 1000):
        j = j**i
except ArithmeticError as e:
    print(f"{e}, {e.__class__}")
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in Python is a base class for exceptions that occur when a lookup or indexing operation fails. It serves as a parent class for more specific lookup-related exceptions like KeyError and IndexError. The LookupError class is useful when you want to catch multiple lookup-related exceptions in a single except block.

```python
#key error
a = {"a" : 1 ,"b" : 2 ,"c" : 3}
try:
    value=a["d"]
    print(value)
except KeyError :
    print("Error: The key does not exist in the dictionary.")
```

```python
#index error
a=[1,2,3,4,5]
try:
    print(a[8])
except Exception as e:
    print(e)
```

Q5. Explain ImportError. What is ModuleNotFoundError?

ImportError is a Python exception that is raised when an import statement fails to locate and import a module or a specific attribute from a module. It occurs when there are issues related to importing modules in your Python code.

```python
try:
    import non_module
except ImportError:
    print("ERROR: No module named 'non_module' ")
```

 ModuleNotFoundError is a subclass of ImportError. It was introduced in Python 3.6 to provide a more specific exception when a module cannot be found during the import process. In Python 3.6 and later versions, when a module is not found, ModuleNotFoundError is raised instead of the generic ImportError

```python
try:
    import non_existent_module
except ModuleNotFoundError:
    print("Error: The module could not be found.")
```
